```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
df = pd.read_csv('insurance_data.csv')
df.head(5)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.age,df['bought_insurance'],color='red')
plt.xlabel('age')
```

```python
inputs = df[['age']]
target = df['bought_insurance']
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
import numpy as np
x_values = np.linspace(df.age.min(), df.age.max(), 100)
y_values = model.predict_proba(x_values.reshape(-1, 1))[:, 1]  # Probability of class 1
```

```python
plt.scatter(df.age, df['bought_insurance'], color='red', label='Data Points')
plt.xlabel('Age')
plt.plot(x_values, y_values, color='blue', label='Logistic Regression Model')
plt.legend()
```
