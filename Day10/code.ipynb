```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
iris.target
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df
```

```python
df['target_names'] = df.target.apply(lambda x: iris.target_names[x])
df
```

```python
df0 = df[df.target==0]
df0.head()
```

```python
df1 = df[df.target==1]
df1.head()
```

```python
df2 = df[df.target==2]
df2.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='red',marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='blue',marker='.')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='red',marker='*')
```

```python
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='red',marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='blue',marker='.')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='red',marker='*')
```

```python
# according to the scatter plot of petal length and petal width we can use decision tree 
inputscol = df.drop(['sepal length (cm)','sepal width (cm)','target','target_names'],axis='columns')
target = df.target
inputscol
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(inputscol,target,test_size=0.3)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test) *100
```
